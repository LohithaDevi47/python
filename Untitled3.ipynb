```python
def count():
    file_path=input("Enter file path:")
    try:
        with open(file_path,'r')as file:
            lines = file.readlines()
        n_l=len(lines)
        n_w=0
        n_c=0
        for line in lines:
            n_c+=len(line)
            n_w+=len(line.split())
        print(f"no.of lines:{n_l}")
        print(f"no.of words:{n_w}")
        print(f"no.of characters:{n_c}")
    except FileNotFoundError:
        print("file is not found")
    except Exception as e:
        print("An error is occured")
count()
```

```python
import pickle
student={"name":"lohi","department":"aiml","college":"acet"}
with open("student.dat","wb")as file:
    pickle.dump(student, file)
print("student info saved successfully")
with open("student.dat","rb")as file:
    loaded_student = pickle.load(file)
print(f"name : {loaded_student['name']}")
print(f"department : {loaded_student['department']}")
print(f"college : {loaded_student['college']}")

```
